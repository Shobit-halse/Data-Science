# 3.16 Applying a Function Over All Elements in a Column

```python
def uppercase(x):
    if isinstance(x, str):
        return x.upper()
    return x
print(df['Aircraft'].apply(uppercase)[0:5])
```

# 3.17 Applying a Function to Groups

```python
df.groupby('Country')['Total fatalities'].apply(lambda x: x.max() - x.min())
```

# 3.18 Concatenating DataFrames

# 3.19 Merging DataFrames

```python
df_employees = pd.DataFrame({
    'employee_id': ['1', '2', '3', '4'],
    'employee_name': ['John', 'Jane', 'Bob', 'Alice']
})

df_sales = pd.DataFrame({
    'employee_id': ['3', '4', '5', '6'],
    'total_sales': [23456, 2512, 2345, 1455]
})

pd.merge(df_employees, df_sales, on='employee_id')
```
